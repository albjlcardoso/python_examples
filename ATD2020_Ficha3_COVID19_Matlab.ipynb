Objetivo: Pretende-se prever a evoluo do novo Coronavrus COVID-19 em Portugal, em funo da evoluo em Itlia e em Espanha.


**Exerccio**
	

Os dados recentes do COVID-19 em Itlia, Espanha e Portugal so:

 H 19 dias a Itlia tinha 21 casos, h 9 dias tinha 2036 e hoje tem 12142

 H 11 dias a Espanha tinha 58 casos e hoje tem 2277

 H 19 dias a Itlia tinha 1 morte, h 9 dias tinha 52 e hoje tem 827.

 H 11 dias a Espanha tinha zero mortes e hoje tem 55.

 Ns hoje temos um nmero semelhante ao que a Espanha tinha h dez dias e a Itlia h 19 dias.

 Como estaremos ns no final de Maro?

```matlab
%--- Dados
n_IT=[0:11 18 19];
x_IT=[21 79 157 229 323 470 655 889 1128 1701 2036 2502 10149 12142];
n_SP=[0:9 11];
x_SP=[58 120 165 228 282 401 525 674 1231 1695 2277];
n_PT=0;
x_PT=59;
```

```matlab
%plot inline -w 1200
%--- Representao grfica das evolues do n de casos de COVID-19

figure(1)
plot(n_IT,x_IT,'-o',n_SP,x_SP,'-+',n_PT,x_PT,'-*')
xlabel('n [dias]')
title('N de casos de COVID-19')
legend('Italy','Spain','Portugal','Location','northwest')
```

```matlab
%---Verificar a correlao entre a evoluo em Itlia e em Espanha
corr_IT_SP=corr(x_IT(1:10)',x_SP(1:10)')
% Tendncia polinomial de grau 2
p=polyfit(n_IT,x_IT,2)
% Tendncia exponencial 
f = fit(n_IT',x_IT','exp1')
```

```matlab
%---Previso da evoluo do COVID-19
n_IT2=[n_IT 21:30];
x_IT2=polyval(p,n_IT2); % polinimial de 2 grau
x_ITe=f.a*exp(f.b*n_IT2); % exponencial

figure(2)
plot(n_IT,x_IT,'-o',n_SP,x_SP,'-+',n_PT,x_PT,'-*',...
    n_IT2,x_IT2,'-d',n_IT2,x_ITe,'-s')
xlabel('n [dias]')
title('N de casos de COVID-19')
legend('Italy','Spain','Portugal',...
    'previso de 2 grau','previso exponencial','Location','northwest')
```
