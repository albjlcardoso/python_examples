Objetivo: Pretende-se prever a evolução do novo Coronavírus COVID-19 em Portugal, em função da evolução em Itália e em Espanha.


**Exercício**
	

Os dados recentes do COVID-19 em Itália, Espanha e Portugal são:

• Há 19 dias a Itália tinha 21 casos, há 9 dias tinha 2036 e hoje tem 12142

• Há 11 dias a Espanha tinha 58 casos e hoje tem 2277

• Há 19 dias a Itália tinha 1 morte, há 9 dias tinha 52 e hoje tem 827.

• Há 11 dias a Espanha tinha zero mortes e hoje tem 55.

• Nós hoje temos um número semelhante ao que a Espanha tinha há dez dias e a Itália há 19 dias.

• Como estaremos nós no final de Março?

```matlab
%--- Dados
n_IT=[0:11 18 19];
x_IT=[21 79 157 229 323 470 655 889 1128 1701 2036 2502 10149 12142];
n_SP=[0:9 11];
x_SP=[58 120 165 228 282 401 525 674 1231 1695 2277];
n_PT=0;
x_PT=59;
```

```matlab
%plot inline -w 1200
%--- Representação gráfica das evoluções do nº de casos de COVID-19

figure(1)
plot(n_IT,x_IT,'-o',n_SP,x_SP,'-+',n_PT,x_PT,'-*')
xlabel('n [dias]')
title('Nº de casos de COVID-19')
legend('Italy','Spain','Portugal','Location','northwest')
```

```matlab
%---Verificar a correlação entre a evolução em Itália e em Espanha
corr_IT_SP=corr(x_IT(1:10)',x_SP(1:10)')
% Tendência polinomial de grau 2
p=polyfit(n_IT,x_IT,2)
% Tendência exponencial 
f = fit(n_IT',x_IT','exp1')
```

```matlab
%---Previsão da evolução do COVID-19
n_IT2=[n_IT 21:30];
x_IT2=polyval(p,n_IT2); % polinimial de 2º grau
x_ITe=f.a*exp(f.b*n_IT2); % exponencial

figure(2)
plot(n_IT,x_IT,'-o',n_SP,x_SP,'-+',n_PT,x_PT,'-*',...
    n_IT2,x_IT2,'-d',n_IT2,x_ITe,'-s')
xlabel('n [dias]')
title('Nº de casos de COVID-19')
legend('Italy','Spain','Portugal',...
    'previsão de 2º grau','previsão exponencial','Location','northwest')
```
