<strong>Exercício 2.</strong> Considere a seguinte função dependente do tempo $f = \sin(2\pi t) + \sin(\pi t)$.

2.1.	Crie um vetor tempo t que permita obter valores para f entre -10 e 10 segundos. Considere uma diferença de 0.01s entre valores do vetor tempo. Indique a dimensão do vetor tempo.

2.2.	Represente graficamente a evolução de f em função do tempo.

```matlab
t=-10:0.01:10;
disp('Dimensao do vetor tempo:')
disp(length(t))

f=sin(2*pi*t)+sin(pi*t);

plot(t,f,'-ob')
xlabel('tempo [s]')
ylabel('f(t)')
title('Grafico de f(t)')
```

<strong>Exercício 3.</strong> Realize o exercício 2, mas recorrendo a cálculo simbólico.

```matlab
syms t  % t=sym('t')
ft=sin(4*pi*t)+sin(pi*t)    % ft=sym('sin(4*pi*t)+sin(pi*t)')
t=-10:0.01:10;
f1=double(subs(ft));
plot(t,f1,'-or')
xlabel('tempo [s]')
ylabel('f(t)')
title('Grafico de f(t)')
```

<strong>Exercício 4.</strong> Crie um script para representar graficamente a função $f(x,y) = \sin(x y) + \cos(x)$ recorrendo a cálculo numérico e a cálculo simbólico.

```matlab
x=-4:0.1:4;  % Definição do domínio de x e de y
y=-4:0.1:4;

[X,Y]=meshgrid(x,y);  % Produz as coordenadas duma grelha retangular (X, Y)

Z=sin(X.*Y)+cos(X);   % Calcula f(x,y)

mesh(X,Y,Z)           % Representação gráfica usando a função mesh
xlabel('x')
ylabel('y')
zlabel('f(x,y)')
title('mesh of f(x,y)=sin(xy)+cos(x)')
```

```matlab
surf(X,Y,Z)      % Representação gráfica usando a função surf
xlabel('x')
ylabel('y')
zlabel('f(x,y)')
title('surf of f(x,y)=sin(xy)+cos(x)')
```

```matlab
syms x y                   % Define x e y como variávies simbólicas
fxy=sin(x*y)+cos(x);       % Calcula f(x,y) simbolicamente
fmesh(fxy,[-4 4 -4 4]);    % Representação gráfica usando a função fmesh
xlabel('x')
ylabel('y')
zlabel('f(x,y)')
title('mesh of symbolic f(x,y)=sin(xy)+cos(x)')
```
